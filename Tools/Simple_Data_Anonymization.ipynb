```python
import numpy as np
import pandas as pd
```

```python
#set masking target column
target = "Name"
```

```python
df = pd.read_csv("train.csv", encoding='utf-8')
df.head()
```

```python
#Replace list
lt = []
abc = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
for i in abc:
    for j in abc:
        for k in abc:
            mask = str(i + j + k)
            lt.append(mask)
```

```python
lt[1]
```

```python
from sklearn.preprocessing import LabelEncoder
#LabelEncoderのインスタンスを生成
le = LabelEncoder()
le = le.fit(df[target]) #set column name
df[target] = le.transform(df[target])
df
```

```python
#Replace
count = 0
for i in df[target].unique():
    df[target] = df[target].replace(i, lt[count])
    count +=1
```
